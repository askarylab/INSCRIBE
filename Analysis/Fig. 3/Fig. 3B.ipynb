```python
import os
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Ensure that the dataframes are defined and contain the necessary data
# Intensity of H2B-CFP for single cells at each CHIR and Dox dosage

# Create a custom color palette with 5 colors
custom_palette = sns.color_palette("cubehelix", 5)

# Filter the DataFrames based on specified conditions
temp_df = Wnt_1_df[(Wnt_1_df['CHIR (uM)'].isin([0, 3])) & 
                   (Wnt_1_df['stimulation time (Days)'].isin([0,1,3,4,5,6,7]))].copy()
temp_df_2 = BMP_1_df[(BMP_1_df['BMP2 (ng/ml )'].isin([0,64])) & 
                     (BMP_1_df['stimulation time (Days)'].isin([0,1,2,3,4,5,6,7]))].copy()
temp_df_3 = BMP_2_df[(BMP_2_df['BMP2 (ng/ml )'].isin([0,64])) & 
                     (BMP_2_df['stimulation time (Days)'].isin([0,1,2,3,4,5,6,7]))].copy()

plt.figure(figsize=(7, 5), dpi=70)

# Custom estimator function to calculate the median
def median_func(y):
    return np.median(y)

# Functions to calculate the 25th and 75th percentiles (IQR)
def iqr_lower(y):
    return np.percentile(y, 25)

def iqr_upper(y):
    return np.percentile(y, 75)

# Plot median points with error bars representing IQR for temp_df
sns.lineplot(x='stimulation time (Days)', 
             y='mout avg mean', 
             data=temp_df, 
             color=custom_palette[4],
             estimator=median_func,
             errorbar=('ci', 95),
             linewidth=0.5,
             markersize=6,
             zorder=1,
             #alpha=0.75,
             label='WNT-BC1')

grouped = temp_df.groupby('stimulation time (Days)')['mout avg mean']
medians = grouped.median()
iqr_lowers = grouped.quantile(0.25)
iqr_uppers = grouped.quantile(0.75)

plt.errorbar(medians.index, medians, yerr=[medians - iqr_lowers, iqr_uppers - medians],
             fmt='o', color=custom_palette[4], capsize=3, capthick=1, ecolor=custom_palette[4])

# Plot median points with error bars representing IQR for temp_df_2
sns.lineplot(x='stimulation time (Days)', 
             y='mout avg mean', 
             data=temp_df_2, 
             color=custom_palette[3],
             estimator=median_func,
             errorbar=('ci', 95),
             linewidth=0.5,
             markersize=6,
             zorder=2,
             label='BMP-BC1')

grouped = temp_df_2.groupby('stimulation time (Days)')['mout avg mean']
medians = grouped.median()
iqr_lowers = grouped.quantile(0.25)
iqr_uppers = grouped.quantile(0.75)

plt.errorbar(medians.index, medians, yerr=[medians - iqr_lowers, iqr_uppers - medians],
             fmt='o', color=custom_palette[3], capsize=3, capthick=1, ecolor=custom_palette[3])

# Plot median points with error bars representing IQR for temp_df_3
sns.lineplot(x='stimulation time (Days)', 
             y='mout avg mean', 
             data=temp_df_3, 
             color=custom_palette[1],
             estimator=median_func,
             errorbar=('ci', 95),
             linewidth=1,
             markersize=6,
             zorder=3,
             label='BMP-BC2')

grouped = temp_df_3.groupby('stimulation time (Days)')['mout avg mean']
medians = grouped.median()
iqr_lowers = grouped.quantile(0.25)
iqr_uppers = grouped.quantile(0.75)

plt.errorbar(medians.index, medians, yerr=[medians - iqr_lowers, iqr_uppers - medians],
             fmt='o', color=custom_palette[1], capsize=3, capthick=1, ecolor=custom_palette[1])

plt.title('Fig. 3B')
plt.xlabel('Recording time (Days)')
plt.ylabel('Average CNN model output of\n barcode arrays in single cell')

#plt.xlim([-0.5, 7.5])
#plt.ylim([-0.5, 12.5])
# Customizing plot appearance
plt.gca().spines['bottom'].set_linewidth(0.625)
plt.gca().spines['left'].set_linewidth(0.625)
plt.gca().spines['top'].set_linewidth(0.625)
plt.gca().spines['right'].set_linewidth(0.625)

# Adding the legend manually
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles=handles, labels=labels, title='Conditions', bbox_to_anchor=(1, 1), loc='upper left')

#save_path = os.path.join(data_dir_3, f'231122_231124_Wnt&BMP_BC1&2_time_dependent_CNN.svg')
#plt.savefig(save_path, dpi=300, format='svg', transparent=True)
plt.show()
```
